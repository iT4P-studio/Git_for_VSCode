**構造体定義 (行 7–10)**

CostFrom 構造体は各頂点に対して、
*   cost：始点からの暫定最短距離を保持し、
*   from：最短経路上でその頂点に到達する直前に辿る「親(前)頂点番号」を記録する。

初期状態では from = -1 と設定し、未到達を示す。（行 20–21）

---

**初期化と始点設定 (行 19–25)**

全ての頂点 i で dist[i].cost = INF、dist[i].from = -1、used[i] = 0 とする。（行 20–23）

始点 start だけ dist[start].cost = 0、dist[start].from = start として、経路のループ起点を自身に設定する。（行 24–25）


---


**未確定頂点の選択 (行 27–36)**

外側のループを iter < N 回回し、毎回行う処理は：

*   最小距離探索 (行 28–34)

dist[i].cost が最小で used[i] == 0 の頂点 v を線形探索により選ぶ。（行 30–32）

*   終了判定 & 確定 (行 35–36)

v < 0 なら未訪問かつ距離が有限な頂点がないためループを早期終了。

そうでなければ used[v] = 1 として頂点 v を確定にする。


---


**緩和 (Relaxation) と経路情報の更新 (行 38–46)**

頂点 v を確定した後、全隣接 u について

adjmat[v][u] != INF かつ used[u] == 0 を満たす場合、

newcost = dist[v].cost + adjmat[v][u] を計算し、

もし newcost < dist[u].cost ならば、

dist[u].cost = newcost と距離を更新し、

dist[u].from = v として「頂点 u の最短経路で直前に来る頂点 v」を記録する。（行 42–43）

この仕組みによって、Dijkstra の実行中に「最短経路の木（Shortest Path Tree）が自然と構築される」。


---


**経路の再帰表示 (行 50–63)**

関数 print_path(dest) は、引数 dest までの最短経路を再帰的に辿って表示する。

*   基本ケースは

dist[dest].from == dest のとき：始点自身なので行番号 53–55 で「頂点番号のみ」を表示。

dist[dest].from == -1 のとき：到達不能なケースで、行番号 55–58 で "no_path" を出力。

*   再帰ケース（行 59–62）：

parent = dist[dest].from が有効な場合、まず print_path(parent) を呼び出して「親→…→parent」を表示し、

その後に " -> dest" を付け加えて経路を連結表示する。

結果として、print_path(i) を呼ぶと「start → … → i」という最短経路ノード列が出力される。


---


**メインの流れ (行 65–91)**

行 67：scanf("%d %d %d", &N, &M, &start); で頂点数 N、辺数 M、開始頂点 start を読み込む。

行 68–74：隣接行列 adjmat を全体初期化。自己ループに 0、その他を INF とする。

行 75–80：各辺 u→v（重み w）を読み込み、「もし重みが従来より小さければ」adjmat[u][v] = w と最小値を格納。

行 82：dijkstra(start) を呼び出して最短距離と経路情報を構築。

行 84–89：全頂点 i に対して print_path(i) を呼んで最短ルートを表示し、続けて (cost=%d) で dist[i].cost を出力。



```
 1  #include <stdio.h>
 2  #include <limits.h>
 3
 4  #define MAXV 100
 5  #define INF   INT_MAX
 6
 7  typedef struct {
 8      int cost;      // (8) 始点からの暫定最短距離を保持
 9      int from;      // (9) 最短経路における直前の頂点を記録
10  } CostFrom;
11
12  int N, M;
13  int adjmat[MAXV][MAXV];     // (13) 隣接行列。存在しない辺は INF
14  CostFrom dist[MAXV];
15  int used[MAXV];
16
17  // (17) Dijkstra 本体：開始頂点 start を受け取り dist[] を更新
18  void dijkstra(int start) {
19      for (int i = 0; i < N; i++) {
20          dist[i].cost = INF;   // (20) 全頂点の距離を無限大で初期化
21          dist[i].from = -1;    // (21) 「親」が未定義であることを示す
22          used[i] = 0;
23      }
24      dist[start].cost = 0;     // (24) 始点は距離 0
25      dist[start].from = start; // (25) 始点の from に自分自身を設定
26
27      for (int iter = 0; iter < N; iter++) {
28          int v = -1, minval = INF;
29          for (int i = 0; i < N; i++) {
30              if (!used[i] && dist[i].cost < minval) {
31                  minval = dist[i].cost;
32                  v = i;         // (32) 最小距離の未確定頂点 v を選ぶ
33              }
34          }
35          if (v < 0) break;        // (35) これ以上伸ばせる頂点がなければ終了
36          used[v] = 1;             // (36) v を確定済みにする
37
38          for (int u = 0; u < N; u++) {
39              if (!used[u] && adjmat[v][u] != INF) {
40                  int newcost = dist[v].cost + adjmat[v][u];
41                  if (newcost < dist[u].cost) {
42                      dist[u].cost = newcost;  // (42) 距離を更新
43                      dist[u].from = v;        // (43) 経路情報として "v を親" に記録
44                  }
45              }
46          }
47      }
48  }
49
50  // (50) 経路を再帰的に表示する関数
51  void print_path(int dest) {
52      if (dist[dest].from == dest) {
53          // (53) 始点の場合、その番号のみ表示
54          printf("%d", dest);
55      } else if (dist[dest].from == -1) {
56          // (56) 到達不能なら "no_path" と表示
57          printf("no_path");
58      } else {
59          // (59) 親ノードを先に表示し、自分自身を "-> dest" として続ける
60          print_path(dist[dest].from);
61          printf(" -> %d", dest);
62      }
63  }
64
65  int main() {
66      int start;
67      scanf("%d %d %d", &N, &M, &start);
68      // (68) 隣接行列初期化。自己ループは 0、他は INF
69      for (int i = 0; i < N; i++) {
70          for (int j = 0; j < N; j++) {
71              if (i == j) adjmat[i][j] = 0;
72              else        adjmat[i][j] = INF;
73          }
74      }
75      // (75) 辺情報を読み込み、最小重みを設定
76      for (int i = 0; i < M; i++) {
77          int u, v, w;
78          scanf("%d %d %d", &u, &v, &w);
79          if (w < adjmat[u][v]) adjmat[u][v] = w;
80      }
81
82      dijkstra(start);           // (82) Dijkstra を実行
83
84      // (84) 各頂点までの最短経路を文字列として表示
85      for (int i = 0; i < N; i++) {
86          printf("path to %d: ", i);
87          print_path(i);
88          printf(" (cost=%d)\n", dist[i].cost);
89      }
90      return 0;
91  }
```

# 必須課題6B Dijkstraプログラムの計算量


DataAlgo_UT(008)_ShortestPath1N.ipynb で示した shortest_diskstra プログラムの時間計算量と空間計算量を，プログラムの構造を参照しつつ議論せよ．

* （ChatGPT等に相談すると授業内での言及と異なる説明をしてくることが多いので鵜呑みにしないこと．これは本授業の課題すべてに共通）  
* 説明に必要なプログラム参照については，以下のコードセルに必要部分**のみ**を張り付けること．  
* 文章による説明は以下のテキストセルを利用せよ．





---


1. 時間計算量
初期化 (行 18–23)

ここで for (i = 0; i < N; i++) のループを一回回しているため、O(N)。

隣接行列の初期化 (行 57–64)

二重ループで i=0..N-1, j=0..N-1 を回しているので、O(N²)。ただし、Dijkstra 本体の計算量とは独立に一度だけ実行される。

Dijkstra のメインループ (行 27–50)

外側ループ (行 28) が iter=0..N-1 と N 回繰り返される。

各イテレーション内に、

最小距離頂点探索 (行 29–36) で未使用の N 頂点を線形走査し、最小 dist[i].cost を探す：O(N)

隣接頂点の緩和 (行 40–49) で頂点 v から全 N 頂点 u を走査し、adjmat[v][u] を使って緩和を行う：O(N)

したがって、1 回のイテレーションにかかる時間は O(N + N) = O(N)。

それを N 回繰り返すため、Dijkstra 全体は O(N × N) = O(N²)。

合計 (時間計算量)

隣接行列初期化：O(N²)

Dijkstra：O(N²)

それ以外の初期化や読み込みは O(N) or O(M) だが、N² の項に比べ小さいため総合すると O(N² + N²) = O(N²)。


---


2. 空間計算量
隣接行列 adjmat[MAXV][MAXV] (行 13)

サイズは N×N 要素で、各要素に整数を格納するため、O(N²) のメモリを消費する。

距離テーブル dist[MAXV] (行 14)

頂点ごとに CostFrom 構造体をひとつずつ持つので、O(N)。

訪問フラグ used[MAXV] (行 15)

頂点数 N の配列を持つので、O(N)。

その他の補助変数

ローカル変数や一時的なスカラー変数（行 28–36 や 40–49 における int v, int newcost など）は O(1)。

合計 (空間計算量)

O(N²)（隣接行列が支配的）。


---


以上より、shortest_dijkstra プログラムは

時間計算量：O(N²)

空間計算量：O(N²)



```
 1  #include <stdio.h>
 2  #include <limits.h>
 3
 4  #define MAXV 100
 5  #define INF   INT_MAX
 6
 7  typedef struct {
 8      int cost;
 9      int from;
10  } CostFrom;
11
12  int N, M;
13  int adjmat[MAXV][MAXV];   // (13) 隣接行列：辺がなければ INF
14  CostFrom dist[MAXV];
15  int used[MAXV];
16
17  void dijkstra(int start) {
18      // (18–23) 初期化ループ：O(N)
19      for (int i = 0; i < N; i++) {
20          dist[i].cost = INF;
21          dist[i].from = -1;
22          used[i] = 0;
23      }
24      dist[start].cost = 0;
25      dist[start].from = start;
26
27      // (27) メインループ：N回繰り返す
28      for (int iter = 0; iter < N; iter++) {
29          // (29–34) 未確定頂点から最小距離を探索するループ：O(N)
30          int v = -1, minval = INF;
31          for (int i = 0; i < N; i++) {
32              if (!used[i] && dist[i].cost < minval) {
33                  minval = dist[i].cost;
34                  v = i;
35              }
36          }
37          if (v < 0) break;
38          used[v] = 1;
39
40          // (40–46) 頂点 v から全頂点 u への緩和ループ：O(N)
41          for (int u = 0; u < N; u++) {
42              if (!used[u] && adjmat[v][u] != INF) {
43                  int newcost = dist[v].cost + adjmat[v][u];
44                  if (newcost < dist[u].cost) {
45                      dist[u].cost = newcost;
46                      dist[u].from = v;
47                  }
48              }
49          }
50      }
51  }
52
53  int main() {
54      int start;
55      scanf("%d %d %d", &N, &M, &start);
56
57      // (57–64) 隣接行列初期化：O(N^2)
58      for (int i = 0; i < N; i++) {
59          for (int j = 0; j < N; j++) {
60              if (i == j) adjmat[i][j] = 0;
61              else        adjmat[i][j] = INF;
62          }
63      }
64      for (int i = 0; i < M; i++) {
65          int u, v, w;
66          scanf("%d %d %d", &u, &v, &w);
67          if (w < adjmat[u][v]) adjmat[u][v] = w;
68      }
69
70      dijkstra(start);  // (70) Dijkstra 呼び出し：O(N^2)
71
72      return 0;
73  }
```

(8–9) dist[][] と via[][] の役割

dist[i][j] に i→j の現在の最短距離を保持し、via[i][j] には「最短経路 i→j を実現する際の最初の中継点 k」を記録する。via[i][j] < 0 の場合は直接 i→j の辺か i==j、あるいは未到達を示す。


---


(11–25) floyd() 関数内の更新

行 17–18: i→k と k→j が到達可能 (dist[i][k] < INF && dist[k][j] < INF) であり、かつ dist[i][k] + dist[k][j] < dist[i][j] を満たすなら、

行 19 で dist[i][j] を i→k→j の新しい距離に更新し、

行 20 で via[i][j] = k と記録することで、「i→j の最短経路は i→…→k→…→j」であることを示す。


---


(27–38) printPath(int i, int j) の再帰表示

行 29–33 (via<0 の場合)

via[i][j] < 0 のとき、i→j は直接の辺（または i==j）。

その際はまず printf("%d", i) で i を出力し、i!=j なら " -> j" を続けることで「i→j」の形式で表示する。

行 34–37 (via≥0 の場合)

int k = via[i][j]; で中継点 k を取り出す。

再帰的に (行 35) printPath(i, k); を呼び出すことで i→k の部分経路を先に出力し、続いて (行 36) printPath(k, j); で k→j の部分経路を出力する。

これにより i と j の間に複数の中継点が存在する場合でも、via[][] の再帰的分割によって「i → … → k → … → j」のすべてのノードを正しい順序で表示できる。

(41–48) 初期化

行 45–46 で i→i を 0、その他を INF にし、最初は全頂点対間で直接経路があるかどうかを表す。

行 47–48 で via[i][j] = -1 として「まだ中継点は設定されていない」ことを示す。


---


(50–57) 辺情報の読み込み

各辺 u→v の重み w を読み込み、もし w が従来 dist[u][v] より小さければ、行 55–56 で dist[u][v] = w とし、経路上に中継を使わない直接辺であることを via[u][v] = -1 と設定する。


---


(59) Floyd–Warshall の呼び出し

ここで全頂点対 (i,j) の最短距離 dist[i][j] と、中継点情報 via[i][j] が最終的に確定する。

(61–68) 最終結果の表示

行 65–67: dist[i][j] == INF の場合は到達不能として "no_path" を表示。

行 68–70: 到達可能な場合は printPath(i,j) で最短経路ノード列を再帰的に表示し、続けて (cost=dist[i][j]) を出力する。

```
 1  #include <stdio.h>
 2  #include <limits.h>
 3
 4  #define MAXV 100
 5  #define INF   INT_MAX
 6
 7  int N, M;
 8  int dist[MAXV][MAXV];  // (8) 頂点i→j間の最短距離を保持
 9  int via[MAXV][MAXV];   // (9) 中継頂点 k を記録。via[i][j]<0 は直通または未定義を示す
10
11  // (11) Floyd–Warshall 本体：dist[][] と via[][] を更新
12  void floyd() {
13      for (int k = 0; k < N; k++) {
14          for (int i = 0; i < N; i++) {
15              for (int j = 0; j < N; j++) {
16                  // (16) i→k および k→j が到達可能かチェック
17                  if (dist[i][k] < INF && dist[k][j] < INF
18                      && dist[i][k] + dist[k][j] < dist[i][j]) {
19                      dist[i][j] = dist[i][k] + dist[k][j];  // (19) 距離更新
20                      via[i][j]  = k;                        // (20) i→j の中継最初は k
21                  }
22              }
23          }
24      }
25  }
26
27  // (27) 経路表示：i→j の最短経路を再帰的に分割して出力
28  void printPath(int i, int j) {
29      if (via[i][j] < 0) {
30          // (30) via<0 なら i→j は直接辺または i==j
31          printf("%d", i);
32          if (i != j) printf(" -> %d", j);
33      } else {
34          int k = via[i][j];             // (34) 中継点 k を取り出す
35          printPath(i, k);               // (35) i→k の部分経路を再帰表示
36          printPath(k, j);               // (36)   k→j の部分経路を再帰表示
37      }
38  }
39
40  int main() {
41      scanf("%d %d", &N, &M);
42      // (41–48) 初期化：自己ループは 0、その他は INF、via=−1
43      for (int i = 0; i < N; i++) {
44          for (int j = 0; j < N; j++) {
45              if (i == j)      dist[i][j] = 0;
46              else             dist[i][j] = INF;
47              via[i][j] = -1;
48          }
49      }
50      // (50–57) 辺情報を読み込み、最小重みを dist[u][v] に登録
51      for (int e = 0; e < M; e++) {
52          int u, v, w;
53          scanf("%d %d %d", &u, &v, &w);
54          if (w < dist[u][v]) {
55              dist[u][v] = w;
56              via[u][v]  = -1;  // 直接辺なので中継なしを示す
57          }
58      }
59      floyd();  // (59) 全頂点対の最短距離と via を計算
60
61      // (61–68) 結果を表示：到達不可は "no_path"
62      for (int i = 0; i < N; i++) {
63          for (int j = 0; j < N; j++) {
64              printf("Path %d->%d: ", i, j);
65              if (dist[i][j] == INF) {
66                  printf("no_path\n");
67              } else {
68                  printPath(i, j);
69                  printf(" (cost=%d)\n", dist[i][j]);
70              }
71          }
72      }
73      return 0;
74  }
```

# 必須課題6D Floyd プログラムの計算量

DataAlgo_UT(009)_ShortestPathNN.ipynb で示した shortest_floyd プログラムの時間計算量と空間計算量を，プログラムの構造を参照しつつ議論せよ．

* 授業内で言及していた方針に従ってまとめること
* （ChatGPT等に相談すると授業内での言及と異なる説明をしてくることが多いので鵜呑みにしないこと．これは本授業の課題すべてに共通）  
* 説明に必要なプログラム参照については，以下のコードセルに必要部分**のみ**を張り付けること．  
* 文章による説明は以下のテキストセルを利用せよ．





---


**時間計算量**

入力と初期化部 (行 27–34, 36–41)：

二重ループで i=0..N-1 と j=0..N-1 を回し、dist[i][j] と via[i][j] を設定する部分は O(N²)。

辺情報の読み込みループ (行 36–41) は M 回の入力処理であり、各回 O(1)。従って合計で O(M)。

これらは Floyd 本体とは独立して一度だけ実行される。

Floyd–Warshall 本体 (行 13–23)：

外側に k = 0..N-1 のループ (行 13)、次に i = 0..N-1 のループ (行 14)、内側に j = 0..N-1 のループ (行 15) がネストしている。

各 (k,i,j) に対して条件チェックと更新処理 (行 16–19) を定数時間で行うため、
3 重ループ全体で O(N × N × N) = O(N³) の反復が行われる。

したがって、Floyd–Warshall の計算時間は O(N³)。

出力部 (行 48–54)：

二重ループで i=0..N-1 と j=0..N-1 を回し、結果を出力する部分も O(N²)。

しかし O(N³) のパートに比べると小さい項なので、総合のオーダーは O(N³) が支配的である。

合計時間計算量：

入力・初期化 O(N² + M)、

Floyd–Warshall O(N³)、

出力 O(N²)

よって O(N³) と表記できる。


---


**空間計算量**

距離テーブル dist[MAXV][MAXV] (行 8)：

N×N の配列を持つため、O(N²) のメモリを消費する。

中継点情報 via[MAXV][MAXV] (行 9)：

こちらも N×N の配列で、中継ノードを保持するため、やはり O(N²) のメモリ。

その他の変数：

スカラー変数やループカウンタ (i, j, k など) は O(1)。

各頂点対への結果を保持する追加テーブルは特にない。

合計空間計算量：

主要な消費は二つの N×N 配列で、合計O(N² + N²) = O(N²)。


---


以上のように、Floyd–Warshall を用いた shortest_floyd プログラムは

時間計算量：O(N³)

空間計算量：O(N²)

```
 1  #include <stdio.h>
 2  #include <limits.h>
 3
 4  #define MAXV 100
 5  #define INF   INT_MAX
 6
 7  int N, M;
 8  int dist[MAXV][MAXV];   // (8) 頂点 i→j 間の距離テーブル
 9  int via[MAXV][MAXV];    // (9) 中継頂点情報
10
11  void floyd() {
12      // (12–24) 三重ループで全頂点対の更新を行う
13      for (int k = 0; k < N; k++) {                // (13)
14          for (int i = 0; i < N; i++) {            // (14)
15              for (int j = 0; j < N; j++) {        // (15)
16                  if (dist[i][k] < INF && dist[k][j] < INF
17                      && dist[i][k] + dist[k][j] < dist[i][j]) {
18                      dist[i][j] = dist[i][k] + dist[k][j];  // (18)
19                      via[i][j]  = k;                       // (19)
20                  }
21              }
22          }
23      }
24  }
25
26  int main() {
27      // (27–34) 入力と初期化：O(N²)
28      scanf("%d %d", &N, &M);
29      for (int i = 0; i < N; i++) {
30          for (int j = 0; j < N; j++) {
31              if (i == j)      dist[i][j] = 0;
32              else             dist[i][j] = INF;
33              via[i][j] = -1;
34          }
35      }
36      for (int e = 0; e < M; e++) {                  // (36–41)
37          int u, v, w;
38          scanf("%d %d %d", &u, &v, &w);
39          if (w < dist[u][v]) {
40              dist[u][v] = w;
41              via[u][v]  = -1;
42          }
43      }
44
45      floyd();                                      // (45)
46
47      // (47–54) 結果表示ループ：O(N²) （省略可）
48      for (int i = 0; i < N; i++) {
49          for (int j = 0; j < N; j++) {
50              if (dist[i][j] == INF) printf("no_path ");
51              else                    printf("%d ", dist[i][j]);
52          }
53          printf("\n");
54      }
55      return 0;
56  }
```

# 必須課題6E Dijkstraのプログラムを再利用したN対N最短経路推定プログラム

DataAlgo_UT(008)_ShortestPath1N.ipynb で示した shortest_diskstra プログラムを改良して，N対Nの最短経路問題を解くプログラムを実現せよ．  
プログラム内で開始頂点を1からNまで変更しながら，dijkstraのアルゴリズムを実行するという方針である．  
改良したプログラムについて，時間計算量と空間計算量を議論せよ．



```
%%writefile graph6.h
// 8 nodes, undirected, no loop, positive weight.
// NC means no edges.
// NC will be treated as "inifinity" on searching the shortest path.
#define N 8
#define NC 9999 // this big value means both no path and infinity
int edge[N][N] = {
//     0   1   2   3   4   5   6   7
	{  0,100, NC, NC,145,200, NC, NC}, // 0
	{100,  0, 23, NC, NC, 80, NC, 70}, // 1
	{ NC, 23,  0, 21, NC, 45, 37, NC}, // 2
	{ NC, NC, 21,  0, NC, NC, 18,  5}, // 3
	{145, NC, NC, NC,  0, 20, NC, NC}, // 4
	{200, 80, 45, NC, 20,  0,  2, NC}, // 5
	{ NC, NC, 37, 18, NC,  2,  0,  1}, // 6
	{ NC, 70, NC,  5, NC, NC,  1,  0}  // 7
};
```

```
%%writefile Repor6E.c
#include <stdio.h>
#include <limits.h>
#include <stdlib.h>
#include "graph6.h"   // N, NC, edge[][] を読み込む

#define INF NC

typedef struct {
    int cost;      // 始点からの暫定最短距離
    int from;      // 最短経路における直前の頂点
} CostFrom;

int adjmat[N][N];            // 隣接行列。edge[][] の NC を INF としてコピー
CostFrom dist[N];
int used[N];
int parent[N][N];            // parent[s][v] : 始点 s から v への最短経路で v の直前の頂点

// Dijkstra を 1 回実行し、result_dist に距離を、parent[s][] に経路情報を格納する関数
void dijkstra(int start, int result_dist[N]) {
    // (1) 初期化：全頂点を未訪問、距離 INF、from=-1 とする
    for (int i = 0; i < N; i++) {
        dist[i].cost = INF;
        dist[i].from = -1;
        used[i] = 0;
    }
    // (2) 始点の距離は 0、自身を from に設定
    dist[start].cost = 0;
    dist[start].from = start;

    // (3) N 回の反復：最短経路を確定していく
    for (int iter = 0; iter < N; iter++) {
        int v = -1, minval = INF;
        // (3-1) 未使用頂点のうち最小距離を持つ頂点 v を線形探索
        for (int i = 0; i < N; i++) {
            if (!used[i] && dist[i].cost < minval) {
                minval = dist[i].cost;
                v = i;
            }
        }
        // (3-2) もし v<0 ならこれ以上伸ばせる頂点がないので終了
        if (v < 0) break;
        // (3-3) 頂点 v を確定済みにする
        used[v] = 1;

        // (3-4) v から全隣接 u への緩和処理
        for (int u = 0; u < N; u++) {
            if (!used[u] && adjmat[v][u] != INF) {
                int newcost = dist[v].cost + adjmat[v][u];
                if (newcost < dist[u].cost) {
                    dist[u].cost = newcost;
                    dist[u].from = v;
                }
            }
        }
    }
    // (4) 結果を呼び出し元テーブルにコピーし、parent も記録
    for (int i = 0; i < N; i++) {
        result_dist[i] = dist[i].cost;
        parent[start][i] = dist[i].from;
    }
}

// 再帰的に経路を表示するヘルパー
void print_path(int s, int v) {
    if (v == s) {
        // 始点に到達したら番号を出力して終了
        printf("%d", s);
    } else if (parent[s][v] < 0) {
        // 到達不能
        printf("no_path");
    } else {
        // まず親を再帰表示し、最後に自分自身を出力
        print_path(s, parent[s][v]);
        printf(" -> %d", v);
    }
}

int main() {
    // (0) graph6.h にて N, NC, edge[][] が定義済み
    //    ここでは adjmat を作成し、NC を INF に変換する
    for (int i = 0; i < N; i++) {
        for (int j = 0; j < N; j++) {
            if (edge[i][j] == NC) {
                adjmat[i][j] = INF;
            } else {
                adjmat[i][j] = edge[i][j];
            }
        }
    }

    // (5) 全頂点対最短距離を格納する 2 次元配列
    int all_dist[N][N];

    // parent 配列を初期化（到達不能を -1 で示す）
    for (int s = 0; s < N; s++) {
        for (int v = 0; v < N; v++) {
            parent[s][v] = -1;
        }
    }

    // (6) 各始点 s = 0..N-1 に対して Dijkstra を実行
    for (int s = 0; s < N; s++) {
        dijkstra(s, all_dist[s]);
    }

    // (7) 結果を出力：始点 s から各頂点 v までのコストと経路を表示
    for (int s = 0; s < N; s++) {
        printf("From %d:\n", s);
        for (int v = 0; v < N; v++) {
            if (s == v) {
                printf(" To %d : cost 0 : %d\n", v, s);
            } else if (all_dist[s][v] == INF) {
                printf(" To %d : no_path\n", v);
            } else {
                printf(" To %d : cost %d : ", v, all_dist[s][v]);
                print_path(s, v);
                printf("\n");
            }
        }
        printf("\n");
    }

    return 0;
}
```

*   時間計算量

1回の Dijkstra（隣接行列＋線形探索版） にかかる時間は O(N²)。

これを N 回繰り返すので、合計で O(N³)。

*   空間計算量

隣接行列 adjmat[N][N]：O(N²)

結果格納用 all_dist[N][N]：O(N²)

Dijkstra 内部の dist[N], used[N]：O(N)

合計で O(N²)。
